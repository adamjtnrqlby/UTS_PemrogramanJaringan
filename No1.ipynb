```python
import socket
import threading

def log(msg):
    print(msg)

def run_server():
    HOST = "127.0.0.1"
    PORT = 5050

    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind((HOST, PORT))
    server_socket.listen(1)

    log("SERVER AKTIF & MENUNGGU KONEKSI...")
    log(f"Port: {PORT}")

    conn, addr = server_socket.accept()
    log(f"Client Terhubung: {addr}")

    while True:
        data = conn.recv(1024)
        if not data:
            break

        log(f"Pesan diterima: {data.decode()}")
        log("Mengirim kembali ke client...")

        conn.sendall(data)

    conn.close()
    server_socket.close()
    log("Server dihentikan")

# Jalankan server di background
thread = threading.Thread(target=run_server)
thread.start()
```

```python
import socket

def cprint(msg):
    print(msg)

HOST = "127.0.0.1"
PORT = 5050

client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
client_socket.connect((HOST, PORT))

pesan = "Tes Koneksi"
cprint(f"Mengirim pesan ke server: '{pesan}'")

client_socket.sendall(pesan.encode())

data = client_socket.recv(1024)
cprint(f"Balasan diterima dari server: '{data.decode()}'")

client_socket.close()
cprint("Client selesai")
```
